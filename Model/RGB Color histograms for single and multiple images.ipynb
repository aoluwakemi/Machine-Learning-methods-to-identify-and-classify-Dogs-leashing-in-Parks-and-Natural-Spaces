# RGB Color histograms for single and multiple images

```python
import numpy as np
import skimage.color
import skimage.io
import matplotlib.pyplot as plt
import glob
%matplotlib widget

image = skimage.io.imread(r"C:\Users\John Taiwo\Desktop\opencv\data\test\looselyChained\lc27.jpg")

# display the image
fig, ax = plt.subplots()
plt.imshow(image)
plt.show()
```

```python
# tuple to select colors of each channel line
colors = ("red", "green", "blue")
channel_ids = (0, 1, 2)

# create the histogram plot, with three lines, one for each color
plt.figure()
plt.xlim([0, 256])
for channel_id, c in zip(channel_ids, colors):
    histogram, bin_edges = np.histogram(
        image[:, :, channel_id], bins=256, range=(0, 256)
    )
    plt.plot(bin_edges[0:-1], histogram, color=c)

plt.title("Single Image RGB Color Histogram")
plt.xlabel("Color value")
plt.ylabel("Pixel count")

plt.show()
```

```python
images_path = glob.glob(r"C:\Users\John Taiwo\Desktop\opencv\data\train\looselyChained\*")
print(len(images_path))
```

```python
plt.figure()
plt.xlim([0, 256])
plt.ylim([0, 6000])

for item in images_path:
    image = skimage.io.imread(item)

    for channel_id, c in zip(channel_ids, colors):
        histogram, bin_edges = np.histogram(
            image[:, :, channel_id], bins=256, range=(0, 256)
        )
        plt.plot(bin_edges[0:-1], histogram, color=c)


plt.title("Multiple Images RGB Color Histogram")
plt.xlabel("Color value")
plt.ylabel("Pixel count")

plt.show()
```
